```python
df1['Gender'] = df1['Gender'].replace('MALE', 'FEMALE')
```

```python
df1['Tumour_Stage'] = df1['Tumour_Stage'].replace({'I': 1, 'II': 2, 'III': 3})
```

```python
df1['PR status'].value_counts()
```

```python
df1['HER2 status'] = df1['HER2 status'].replace({'Negative': 0, 'Positive': 1})
```

```python
df1[['Day_of_Surgery', 'Month_of_Surgery', 'Year_of_Surgery']] = df1['Date_of_Surgery'].str.split('-', expand=True)
```

```python
df1[['Day_of_Last_Visit', 'Month_of_Last_Visit','Year_of_Last_Visit']] = df1['Date_of_Last_Visit'].str.split('-', expand=True)
```

```python
df1['Month_of_Surgery'] = df1['Month_of_Surgery'].replace({'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12 })
```

```python
df1['Month_of_Last_Visit'] = df1['Month_of_Last_Visit'].replace({'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12 })
```

```python
df1=df1.drop(columns=['Date_of_Surgery'],axis=1)
df1=df1.drop(columns=['Date_of_Last_Visit'],axis=1)
```

```python
df1['Day_of_Surgery'] = df1['Day_of_Surgery'].astype(int)
df1['Year_of_Surgery'] = df1['Year_of_Surgery'].astype(int)
df1['Day_of_Last_Visit'] = df1['Day_of_Last_Visit'].astype(int)
df1['Year_of_Last_Visit'] = df1['Year_of_Last_Visit'].astype(int)
```
